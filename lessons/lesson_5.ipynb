```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  
```

```python
# Чтение набора данных
df = pd.read_csv("datasets/HeightWeight.csv")

# pd.read_excel("file.xlsx", sheet_name="Sheet1")
# pd.read_json("data.json")
# pd.read_csv("file.csv", sep=";", encoding="utf-8")
# df = pd.read_csv("https://raw.githubusercontent.com/.../file.csv")
# import sqlalchemy
# engine = sqlalchemy.create_engine("sqlite:///mydb.db")
# df = pd.read_sql("SELECT * FROM table", engine)
```

```python
df.head()
```

```python
df = df.rename(columns = {"Height(Inches)": "height", "Weight(Pounds)": "weight"})
```

```python
df.isna().sum()
```

```python
sns.histplot(df["height"], kde = True)
```

```python
sns.kdeplot(df["height"])
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

sns.histplot(df['height'],kde = True, ax=ax)

ax.set_title("Распределение роста")
ax.set_xlabel("Рост, дюймы")
ax.set_ylabel("Частота")
plt.savefig("../images/height_distribution.png")
plt.show()
```

```python
mu = df['height'].mean()
med = df['height'].median()

fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(df['height'], kde=True, ax=ax, color= "blue")
ax.axvline(mu, color="red", linestyle="--", label=f"mean = {mu:.2f}")
ax.axvline(med, color="green", linestyle="-.", label=f"median = {med:.2f}")
ax.set_title("Распределение роста с mean/median")
ax.set_xlabel("Рост, дюймы")
ax.set_ylabel("Частота")
ax.legend()
plt.show()
```
