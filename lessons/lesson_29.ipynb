# Р’РІРөРҙРөРҪРёРө РІ РҪРөР№СҖРҫРҪРҪСӢРө СҒРөСӮРё

````python
import numpy as np
import matplotlib.pyplot as plt
````

## РҹРҫР»РҪСӢР№ СҶРёРәР» РҫРұСғСҮРөРҪРёСҸ

### РЁР°Рі 1: РӨСғРҪРәСҶРёСҸ Р°РәСӮРёРІР°СҶРёРё - Sigmoid

````python
def sigmoid(x):
    return 1/(1+np.exp(-x))
````

### РЁР°Рі 2: Р”Р°РҪРҪСӢРө РҙР»СҸ РұРёРҪР°СҖРҪРҫР№ РәР»Р°СҒСҒРёС„РёРәР°СҶРёРё

````python
# X < 2.5 -> РәР»Р°СҒСҒ 0
# X >=2.5 -> РәР»Р°СҒСҒ 1
X = np.array([1,2,3,4], dtype = np.float32)
Y = np.array([0,0,1,1], dtype = np.float32)
````

### РЁР°Рі 3: Forward Pass - РҪРөР№СҖРҫРҪ СҒ Р°РәСӮРёРІР°СҶРёРөР№

````python
b = 0.0
w = 0.0
#forward pass РҙР»СҸ РҪРөР№СҖРҫРҪР°
def forward(x):
    return sigmoid(w*x+b)

print("РҹСҖРөРҙСҒРәР°Р·Р°РҪРёСҸ РҙРҫ РҫРұСғСҮРөРҪРёСҸ (СҒ РҫРҙРҪРёРј Р·РҪР°СҮРөРҪРёРөРј)")
pred = forward(X[0])
print(pred)
````

### РЁР°Рі 4: Loss Function - Binary Cross-Entropy

```
Loss = -1/n * ОЈ(y*log(Е·) + (1-y)*log(1-Е·))
```


````python
def loss(y_true,y_pred):
    return -np.mean(y_true * np.log(y_pred) + (1-y_true)* np.log(1-y_pred))    
````

### РЁР°Рі 5: Gradient 

````python
def gradient(x,y_true, y_pred):
    error = y_pred - y_true
    dw = np.mean(x*error)
    db = np.mean(error)
    return dw, db
y_pred = forward(X)
dw, db = gradient(X,Y,y_pred)
print(dw,db)
````

### РЁР°Рі 6: РһРұСғСҮРөРҪРёРө - РҝРҫР»РҪСӢР№ СҶРёРәР»

````python
w = 0.0
b = 0.0
learning_rate = 0.5
n_iteration = 100
loss_history = []
for epoch in range(n_iteration):
    # 1. Forward pass
    y_pred = forward(X)
    # 2. Loss
    l = loss(Y,y_pred)
    loss_history.append(l)
    # 3. Gradients
    dw, db = gradient(X,Y,y_pred)
    # 4. Update weights
    w -= learning_rate * dw
    b-= learning_rate * db
    print(f"РӯРҝРҫС…Р°: {epoch+1}, w = {w:.2f}, b = {b:.2f}, РҫСҲРёРұРәР°: {l:.2f}")
````

## РҳСҒРҝРҫР»СҢР·СғРөРј TensorFlow/Keras

````python
# pip install tensorflow
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

print(f"TensorFlow version: {tf.__version__}")
````

### РҹСҖРёРјРөСҖ 1: 

````python
X_tf = np.array([1, 2, 3, 4], dtype=np.float32).reshape(-1, 1)
Y_tf = np.array([2, 4, 6, 8], dtype=np.float32).reshape(-1, 1)

model_linear = keras.Sequential([
    layers.Dense(1, input_shape=(1,), use_bias=False)  
])

model_linear.compile(optimizer='sgd', loss='mse')
````

````python
predictions = model_linear.predict(X_tf)
print(predictions)
````

### РҹСҖРёРјРөСҖ 2: 

````python
X = np.array([1, 2, 3, 4], dtype=np.float32).reshape(-1, 1)
Y = np.array([0, 0, 1, 1], dtype=np.float32).reshape(-1, 1)

model_nn = keras.Sequential([
    layers.Dense(1, activation='sigmoid', input_shape=(1,))
])

model_nn.compile(
    optimizer=keras.optimizers.SGD(learning_rate=1.0),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model_nn.fit(
    X, Y,
    epochs=1000,
    verbose=0
)

loss_tf, acc_tf = model_nn.evaluate(X, Y, verbose=0)
print(f"\nРўРҫСҮРҪРҫСҒСӮСҢ TensorFlow РјРҫРҙРөР»Рё: {acc_tf * 100:.1f}%")
print(f"Loss: {loss_tf:.4f}")
````

### РҹСҖРёРјРөСҖ 3: Р‘РҫР»РөРө СҒР»РҫР¶РҪР°СҸ СҒРөСӮСҢ РҪР° СҖРөР°Р»СҢРҪСӢС… РҙР°РҪРҪСӢС… - Titanic

РҳСҒРҝРҫР»СҢР·СғРөРј РҙР°СӮР°СҒРөСӮ Titanic РҙР»СҸ РҝСҖРөРҙСҒРәР°Р·Р°РҪРёСҸ РІСӢР¶РёРІР°РөРјРҫСҒСӮРё РҝР°СҒСҒР°Р¶РёСҖРҫРІ.

````python
import pandas as pd

titanic_df = pd.read_csv('datasets/Titanic_Dataset.csv')
titanic_df.head()
````

````python
features = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
target = 'Survived'

data = titanic_df[features + [target]].copy()
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Fare'].fillna(data['Fare'].median(), inplace=True)

data.dropna(inplace=True)
````

````python
X_titanic = data[features].values
y_titanic = data[target].values

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_titanic_scaled = scaler.fit_transform(X_titanic)
````

````python
model_deep = keras.Sequential([
    layers.Dense(16, activation='relu', input_shape=(5,)),  # РЎРәСҖСӢСӮСӢР№ СҒР»РҫР№ 1 (5 РҝСҖРёР·РҪР°РәРҫРІ)
    layers.Dense(8, activation='relu'),                      # РЎРәСҖСӢСӮСӢР№ СҒР»РҫР№ 2
    layers.Dense(4, activation='relu'),                      # РЎРәСҖСӢСӮСӢР№ СҒР»РҫР№ 3
    layers.Dense(1, activation='sigmoid')                    # Р’СӢС…РҫРҙРҪРҫР№ СҒР»РҫР№
])

model_deep.compile(
    optimizer='adam',  
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model_deep.summary()
````

````python
history = model_deep.fit(
    X_titanic_scaled, y_titanic,
    epochs=100,
    batch_size=32,
    verbose=1
)
loss_deep, acc_deep = model_deep.evaluate(X_titanic_scaled, y_titanic, verbose=0)
print(f"Loss: {loss_deep:.4f}")

print(f"\nРўРҫСҮРҪРҫСҒСӮСҢ РҪР° РІСҒРөС… РҙР°РҪРҪСӢС…: {acc_deep * 100:.1f}%")
````
