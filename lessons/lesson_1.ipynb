```python
from functools import reduce

nums = [1, 2, 3, 4, 5, 6]

doubled = list(map(lambda x: x * 2, nums))
print("Умножаем на 2 каждый элемент:", doubled)

even = list(filter(lambda x: x % 2 == 0, nums))
print("Фильтруем каждый элемент", even)

product = reduce(lambda x, y: x * y, nums)
print("Произведение всех чисел:", product)
```

```python
nums = [1, 2, 3, 4]

def make_squares(nums):
    result = []
    for x in nums:
        result.append(x ** 2)
    return result

print("List (def):", make_squares(nums))
print("List comprehension:", [x ** 2 for x in nums])

def make_square_dict(nums):
    result = {}
    for x in nums:
        result[x] = x ** 2
    return result

print("Dict (def):", make_square_dict(nums))
print("Dict comprehension:", {x: x ** 2 for x in nums})

def make_mod2_set(nums):
    result = set()
    for x in nums:
        result.add(x % 2)
    return result

print("Set (def):", make_mod2_set(nums))
print("Set comprehension:", {x % 2 for x in nums})
```

# Задача 1. Анализ оценок студентов

У нас есть список оценок студентов. Нужно:
1) Посчитать средний балл.
2) Отфильтровать тех, кто сдал (>= 70).
3) Найти максимальный и минимальный балл.
4) Создать список оценок переведенных в 10 бальную систему.
5) Произведение всех оценок выше среднего значения

```python
from functools import reduce
scores = [6, 4, 91, 74, 47, 88, 100, 59]
# 1
average = (lambda x: sum(x)/len(x))(scores)
print("Среднее значение: ", average)

# 2
passed = list(filter(lambda x: x >= 70, scores))
print(passed)


# 3
print("Max/min: ", max(passed), min(passed))

# 4
new_system_scores = list(map(lambda x: round(x/10), scores))
print(new_system_scores)

# 5
higher_than_average = list(filter(lambda x: x>=average, scores))
print(higher_than_average)
print("Произведение чисел: ", reduce(lambda x,y: x*y, higher_than_average))
```

# Задача 2. Счётчик для файлов

Написать функцию счётчика для файлов, которая:
1) Принимает имя файла слов
2) Считает количество строк, слов и символов
3) Возвращает результат в виде словаря

```python
text = """Data Analysis - это процесс преобразования данных
в полезные инсайты, которые помогают принимать решения."""

def counting_files(filename):
  results = {'lines': 0,'words': 0, 'chars': 0}
  with open(filename, 'r') as f:
    for line in f:
      results['lines'] += 1
      results['words'] += len(line.split())
      results['chars'] += len(line)
  return results

filename = 'exampleCount.txt'
with open(filename, 'x') as f:
  f.write(text)

print(counting_files(filename))
print(type(counting_files(filename)))
```
